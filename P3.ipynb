### Import packagesimport csv

```python
import csv
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

### Read original image

```python
path = "../CarND-Behavioral-Cloning-P3-Dataset/udacity_data/IMG/"
file_name = "left_2016_12_01_13_34_52_223.jpg"
current_path = path + file_name
image = np.array(Image.open(current_path))
plt.imshow(image)
```

```python
### Camera image Merge
```

```python
path = "../CarND-Behavioral-Cloning-P3-Dataset/udacity_data/IMG/"

center_file_name = "center_2016_12_01_13_34_52_223.jpg"
right_file_name = "right_2016_12_01_13_34_52_223.jpg"
left_file_name = "left_2016_12_01_13_34_52_223.jpg"

center_image = np.array(Image.open(path + center_file_name))
right_image  = np.array(Image.open(path + right_file_name))
left_image   = np.array(Image.open(path + left_file_name))

# Visualize undistortion
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.2, hspace=0.1)

ax1.imshow(left_image)
ax1.set_title('Left Camera Image', fontsize=15)


ax2.set_title('Center Camera Image', fontsize=15)
ax2.imshow(center_image)

ax3.set_title('Right Camera Image', fontsize=15)
ax3.imshow(right_image)



# Save the result as image
plt.savefig(r"./output_images/camera_image_sample.jpg")

```

### Flip the image

```python
flipped_image = cv2.flip(image,1)

# Visualize undistortion
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.2, hspace=0.1)
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=20)
ax2.imshow(flipped_image)
ax2.set_title('Flipped Image', fontsize=20)

# Save the result as image
plt.savefig(r"./output_images/flipped_image_sample.jpg")
```
